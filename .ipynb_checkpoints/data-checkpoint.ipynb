```python
import pandas as pd

df1 = pd.read_csv('bcsc_risk_factors_summarized2_092020.csv')
df2 = pd.read_csv('bcsc_risk_factors_summarized3_092020.csv')
df3 = pd.read_csv('bcsc_risk_factors_summarized1_092020.csv')
```

```python
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
print("Total rows:", len(combined_df))

combined_df.to_csv('bcsc_risk_factors.csv', index=False)
```

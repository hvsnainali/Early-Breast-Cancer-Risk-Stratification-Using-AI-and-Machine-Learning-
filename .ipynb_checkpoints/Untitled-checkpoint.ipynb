```python
df = df[df["cancer"] != 9]
breast = df['breast'].astype('int64')
df['breast_dense_3lvl'] = (
    breast.replace({1:1, 2:1, 3:2, 4:2})
          .where(~breast.isin([9]), 9)
).astype('int64')
df['breast_dense_binary'] = df['breast_dense_3lvl'].replace({1:0, 2:1}).astype('int64')
```

```python
# Age wide bins + midpoint
age = df['age'].astype('Int64')
age_wide_map = {1:1,2:1,3:1, 4:2,5:2, 6:3,7:3, 8:4,9:4, 10:5,11:5, 12:6,13:6}
age_wide_labels = {1:'18-39',2:'40-49',3:'50-59',4:'60-69',5:'70-79',6:'80+'}
df['age_widebin_6lvl'] = age.map(age_wide_map).astype('int64')
df['age_widebin_6lvl_label'] = df['age_widebin_6lvl'].map(age_wide_labels)

age_mid = {1:23.5,2:32.0,3:37.0,4:42.0,5:47.0,6:52.0,7:57.0,8:62.0,9:67.0,10:72.0,11:77.0,12:82.0,13:87.0}
df['age_proxy'] = age.map(age_mid).astype(float)

# BMI midpoint proxy
bmi = df['bmi'].astype('int64')
bmi_mid = {1:22.5, 2:27.5, 3:32.5, 4:37.5, 9:np.nan}
df['bmi_proxy'] = bmi.map(bmi_mid).astype(float)
```

```python
df['menopause'].unique()
```

```python
df['postmeno_dense'] = np.where(
    (df['menopause'] == 2) & (df['breast_dense_binary'] == 1), 1,
    np.where(
        df['menopause'].isin([1,2]) & df['breast_dense_binary'].isin([0,1]), 0, np.nan
    )
).astype('int64')
```

```python
fh = df['family_history'].astype('int64')
bx = df['biopsy'].astype('int64')
df['fam_or_biopsy'] = np.where(
    (fh == 1) | (bx == 1), 1,
    np.where(fh.isin([0,1]) & bx.isin([0,1]), 0, np.nan)
).astype('int64')
```

```python
# Simple polynomials
for col in ['year','age_proxy','bmi_proxy']:
    if col in df.columns:
        df[f'{col}_sq'] = df[col]**2
if set(['year','age_proxy']).issubset(df.columns):
    df['year_x_age'] = df['year'] * df['age_proxy']
if set(['year','bmi_proxy']).issubset(df.columns):
    df['year_x_bmi'] = df['year'] * df['bmi_proxy']
if set(['age_proxy','bmi_proxy']).issubset(df.columns):
    df['age_x_bmi'] = df['age_proxy'] * df['bmi_proxy']

    
# ---------------------------------------------------------------------------
print( " ✓ Engineered columns added ✓ " )
display(df.filter([
    'breast','breast_dense_3lvl','breast_dense_binary',
    'menopause_3lvl','menopause_label',
    'age','age_widebin_6lvl','age_widebin_6lvl_label','age_proxy',
    'bmi','bmi_proxy','postmeno_dense','fam_or_biopsy'
]).sample(8))

# ---------------------------------------------------------------------------
```

## Train Test SPLIT               

```python
X = df[feat_cols].copy()
y = df["cancer"].astype(int)
w = df["count"]  # population weights
```

```python
# Train / Test split (clean test set)
X_tr_all, X_te, y_tr_all, y_te, w_tr_all, w_te = train_test_split(
    X, y, w, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Inner Train / Val split 
X_tr, X_val, y_tr, y_val, w_tr, w_val = train_test_split(
    X_tr_all, y_tr_all, w_tr_all, test_size=0.25, random_state=42, stratify=y_tr_all
)
```

#### preprocessing

```python
# one-hot categoricals; median-impute numerics
pre = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), [c for c in cat_cols if c in X.columns]),
    ("num", SimpleImputer(strategy="median", add_indicator=True),       [c for c in num_cols if c in X.columns]),
])
```

```python
pre.fit(X_tr)
Xt_tr  = pre.transform(X_tr).astype(np.float32)
Xt_val = pre.transform(X_val).astype(np.float32)

# keep full-train arrays to refit the final model later
pre.fit(X_tr_all)  # fit on all training now (once)
Xt_tr_all = pre.transform(X_tr_all).astype(np.float32)
Xt_te     = pre.transform(X_te).astype(np.float32)
```

```python
# small, sensible grid (12 combos)
grid = [
    dict(learning_rate=lr, max_leaf_nodes=mln, min_samples_leaf=msl,
         l2_regularization=l2, max_iter=600, random_state=42)
    for lr  in [0.05, 0.10]
    for mln in [31, 63]
    for msl in [50, 100]
    for l2  in [0.0, 0.1]
]

best_ap, best_params = -1.0, None
results = []
```

```python
for p in grid:
    gb = HistGradientBoostingClassifier(**p)
    gb.fit(Xt_tr, y_tr, sample_weight=w_tr)                
    proba_val = gb.predict_proba(Xt_val)[:,1]

    ap  = average_precision_score(y_val, proba_val, sample_weight=w_val)
    auc = roc_auc_score(y_val, proba_val, sample_weight=w_val)
    results.append((p, ap, auc))
    if ap > best_ap:
        best_ap, best_params = ap, p

print("Top by validation PR-AUC:")
for p, ap, auc in sorted(results, key=lambda t: t[1], reverse=True)[:5]:
    print(f"AP={ap:.3f} | ROC-AUC={auc:.3f} | {p}")

print("\nBest params:", best_params)
```

```python
from sklearn.inspection import permutation_importance
feat_names = pre.get_feature_names_out()

if hasattr(gb_best, "feature_importances_"):
    imp = gb_best.feature_importances_
else:
    # fallback permutation importance on a sample
    rng = np.random.RandomState(42)
    n = Xt_te.shape[0]
    take = min(50000, n)            
    idx = rng.choice(n, size=take, replace=False)

    r = permutation_importance(
        gb_best, Xt_te[idx], y_te.iloc[idx],
        n_repeats=3, random_state=42, scoring="roc_auc"
    )
    imp = r.importances_mean

if len(feat_names) != len(imp):
    feat_names = [f"feat_{i}" for i in range(len(imp))]

top = (pd.DataFrame({"feature": feat_names, "importance": imp})
         .sort_values("importance", ascending=False)
         .head(30))

print("Top features:")
display(top)
```

## 8. Feature lists (from your project config)

```python

# CATEGORICAL codes
CATEGORICAL_COLS = ["race","family_history","menarche","first_birth",
                    "menopause","breast","bmi","biopsy","hormone_therapy","age_widebin_6lvl","breast_dense_3lvl"]

# NUMERIC features 
NUMERIC_COLS = ["year","age_proxy","bmi_proxy",
                "year_sq","age_proxy_sq","bmi_proxy_sq",
                "year_x_age","year_x_bmi","age_x_bmi",
                "biopsy_yes","biopsy_unk",
                "breast_dense_binary","postmeno_dense","fam_or_biopsy"]

# Only keep features that actually exist in df (silent drop for sample data)
CATEGORICAL_COLS = [c for c in CATEGORICAL_COLS if c in df.columns]
NUMERIC_COLS = [c for c in NUMERIC_COLS if c in df.columns]
print("Using", len(NUMERIC_COLS), "numeric and", len(CATEGORICAL_COLS), "categorical columns.")
```

## 9. Split & Preprocess (ColumnTransformer inside pipeline)

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline as SkPipeline

X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)

num_tf = SkPipeline([('imputer', SimpleImputer(strategy='median')),
                     ('scaler', StandardScaler())])
cat_tf = SkPipeline([('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
                     ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])

preprocessor = ColumnTransformer([
    ('num', num_tf, NUMERIC_COLS),
    ('cat', cat_tf, CATEGORICAL_COLS)
], remainder='drop')

print("Train:", X_train.shape, "| Test:", X_test.shape)
```

## 10. Baselines (CV, optimize PR‑AUC)

```python

from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.pipeline import Pipeline as SkPipeline
from sklearn.metrics import average_precision_score
import pandas as pd

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

def cv_pr_auc(model):
    pipe = SkPipeline([('preprocess', preprocessor), ('clf', model)])
    scores = cross_val_score(pipe, X_train, y_train, scoring='average_precision', cv=cv, n_jobs=-1)
    return scores.mean(), scores.std()

rows = []
rows.append(('Dummy(stratified)',) + cv_pr_auc(DummyClassifier(strategy='stratified')))
rows.append(('LogReg(balanced)',) + cv_pr_auc(LogisticRegression(max_iter=200, class_weight='balanced', n_jobs=-1)))
rows.append(('RandomForest(balanced)',) + cv_pr_auc(RandomForestClassifier(n_estimators=400, class_weight='balanced', n_jobs=-1, random_state=RANDOM_STATE)))
rows.append(('GradBoost',) + cv_pr_auc(GradientBoostingClassifier(random_state=RANDOM_STATE)))

try:
    from xgboost import XGBClassifier
    xgb = XGBClassifier(objective='binary:logistic', n_estimators=400, learning_rate=0.05,
                        max_depth=5, subsample=0.8, colsample_bytree=0.8,
                        eval_metric='logloss', n_jobs=-1, random_state=RANDOM_STATE)
    rows.append(('XGBoost',) + cv_pr_auc(xgb))
except Exception as e:
    print("XGBoost unavailable:", e)

baseline_df = pd.DataFrame(rows, columns=['Model','CV_PR_AUC_mean','CV_PR_AUC_std']).sort_values('CV_PR_AUC_mean', ascending=False)
display(baseline_df)
```

## 11. Imbalance strategies (SMOTE inside CV pipeline)

```python

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=RANDOM_STATE, n_jobs=-1)

def cv_pr_auc_smote(model):
    pipe = ImbPipeline([('preprocess', preprocessor), ('smote', smote), ('clf', model)])
    scores = cross_val_score(pipe, X_train, y_train, scoring='average_precision', cv=cv, n_jobs=-1)
    return scores.mean(), scores.std()

smote_rows = []
smote_rows.append(('LogReg+SMOTE',) + cv_pr_auc_smote(LogisticRegression(max_iter=200, n_jobs=-1)))
smote_rows.append(('RF+SMOTE',) + cv_pr_auc_smote(RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=RANDOM_STATE)))
smote_rows.append(('GB+SMOTE',) + cv_pr_auc_smote(GradientBoostingClassifier(random_state=RANDOM_STATE)))
try:
    smote_rows.append(('XGB+SMOTE',) + cv_pr_auc_smote(XGBClassifier(objective='binary:logistic', n_estimators=600, learning_rate=0.05,
                                                                     max_depth=5, subsample=0.8, colsample_bytree=0.8,
                                                                     eval_metric='logloss', n_jobs=-1, random_state=RANDOM_STATE)))
except Exception:
    pass

import pandas as pd
smote_df = pd.DataFrame(smote_rows, columns=['Model','CV_PR_AUC_mean','CV_PR_AUC_std']).sort_values('CV_PR_AUC_mean', ascending=False)
display(smote_df)
```

## 12. Tune XGBoost (optimize PR‑AUC)

```python

from sklearn.model_selection import RandomizedSearchCV
xgb_base = XGBClassifier(objective="binary:logistic", eval_metric="logloss", n_jobs=-1, random_state=RANDOM_STATE)
pipe = ImbPipeline([('preprocess', preprocessor), ('smote', smote), ('clf', xgb_base)])

param_distributions = {
    "clf__n_estimators": [300, 400, 600, 800],
    "clf__max_depth": [3, 4, 5, 6],
    "clf__learning_rate": [0.01, 0.03, 0.05, 0.1],
    "clf__subsample": [0.6, 0.8, 1.0],
    "clf__colsample_bytree": [0.6, 0.8, 1.0],
    "clf__reg_lambda": [0.5, 1.0, 2.0, 5.0],
    "clf__reg_alpha": [0.0, 0.5, 1.0]
}
search = RandomizedSearchCV(pipe, param_distributions, n_iter=30, scoring="average_precision",
                            cv=cv, n_jobs=-1, random_state=RANDOM_STATE, verbose=1, refit=True)
search.fit(X_train, y_train)
best_model = search.best_estimator_
print("Best params:", search.best_params_)
print("CV best PR-AUC:", round(search.best_score_, 6))
```

## 13. Hold‑out probabilities & curves

```python

from sklearn.metrics import average_precision_score, roc_auc_score, precision_recall_curve, roc_curve
y_proba_test = best_model.predict_proba(X_test)[:,1]
ap = average_precision_score(y_test, y_proba_test)
roc = roc_auc_score(y_test, y_proba_test)
print("Holdout PR-AUC:", round(ap,6), "| ROC-AUC:", round(roc,6))
```

```python

# PR curve
import matplotlib.pyplot as plt
prec, rec, thr = precision_recall_curve(y_test, y_proba_test)
plt.figure()
plt.plot(rec, prec, lw=2)
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title(f"Precision-Recall (AP={ap:.3f})")
plt.grid(True); plt.show()
```

```python

# ROC curve
fpr, tpr, _ = roc_curve(y_test, y_proba_test)
import matplotlib.pyplot as plt
plt.figure()
plt.plot(fpr, tpr, lw=2, label=f"AUC={roc:.3f}")
plt.plot([0,1],[0,1],'--', lw=1)
plt.xlabel("FPR"); plt.ylabel("TPR")
plt.title("ROC Curve"); plt.legend(); plt.grid(True); plt.show()
```

## 14. Threshold selection (F1‑optimal) & confusion matrix

```python

import numpy as np
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score
f1 = 2*(prec*rec)/(prec+rec+1e-12)
i = int(np.nanargmax(f1))
best_thresh = thr[i] if i < len(thr) else 0.5
y_pred = (y_proba_test >= best_thresh).astype(int)
print("Best threshold by F1:", round(float(best_thresh),4))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Precision:", round(precision_score(y_test, y_pred, zero_division=0),6),
      "| Recall:", round(recall_score(y_test, y_pred, zero_division=0),6),
      "| F1:", round(f1[i],6))
```

```python
df = df[df["cancer"] != 9]
breast = df['breast'].astype('int64')
df['breast_dense_3lvl'] = (
    breast.replace({1:1, 2:1, 3:2, 4:2})
          .where(~breast.isin([9]), 9)
).astype('int64')
df['breast_dense_binary'] = df['breast_dense_3lvl'].replace({1:0, 2:1}).astype('int64')
```

```python
# Age wide bins + midpoint
age = df['age'].astype('Int64')
age_wide_map = {1:1,2:1,3:1, 4:2,5:2, 6:3,7:3, 8:4,9:4, 10:5,11:5, 12:6,13:6}
age_wide_labels = {1:'18-39',2:'40-49',3:'50-59',4:'60-69',5:'70-79',6:'80+'}
df['age_widebin_6lvl'] = age.map(age_wide_map).astype('int64')
df['age_widebin_6lvl_label'] = df['age_widebin_6lvl'].map(age_wide_labels)

age_mid = {1:23.5,2:32.0,3:37.0,4:42.0,5:47.0,6:52.0,7:57.0,8:62.0,9:67.0,10:72.0,11:77.0,12:82.0,13:87.0}
df['age_proxy'] = age.map(age_mid).astype(float)

# BMI midpoint proxy
bmi = df['bmi'].astype('int64')
bmi_mid = {1:22.5, 2:27.5, 3:32.5, 4:37.5, 9:np.nan}
df['bmi_proxy'] = bmi.map(bmi_mid).astype(float)
```

```python
df['postmeno_dense'] = np.where(
    (df['menopause'] == 2) & (df['breast_dense_binary'] == 1), 1,
    np.where(
        df['menopause'].isin([1,2]) & df['breast_dense_binary'].isin([0,1]), 0, np.nan
    )
).astype('int64')
```

```python
fh = df['family_history'].astype('int64')
bx = df['biopsy'].astype('int64')
df['fam_or_biopsy'] = np.where(
    (fh == 1) | (bx == 1), 1,
    np.where(fh.isin([0,1]) & bx.isin([0,1]), 0, np.nan)
).astype('int64')

# Estrogen exposure window (longer -> higher risk)
df["age_menarche_gap"] = (df["age"] - df["menarche"]).clip(lower=0)

# Nulliparity & years since first birth (if 'first_birth' is age at first live birth; 0/neg = none)
df["nulliparous"] = (df["first_birth"] <= 0).astype(int)
df["years_since_first_birth"] = np.where(df["first_birth"] > 0,
                                         (df["age_proxy"] - df["first_birth"]).clip(lower=0),
                                         np.nan)
# Density interactions
df["density_x_famhx"] = df["breast_dense_binary"] * df["family_history"]
df["density_x_postmeno"] = df["breast_dense_binary"] * (df["menopause"] == 1)

# Missingness is information
df["is_bmi_missing"] = df["bmi_proxy"].isna().astype(int)
```

```python
# Simple polynomials
for col in ['year','age_proxy','bmi_proxy']:
    if col in df.columns:
        df[f'{col}_sq'] = df[col]**2
if set(['year','age_proxy']).issubset(df.columns):
    df['year_x_age'] = df['year'] * df['age_proxy']
if set(['year','bmi_proxy']).issubset(df.columns):
    df['year_x_bmi'] = df['year'] * df['bmi_proxy']
if set(['age_proxy','bmi_proxy']).issubset(df.columns):
    df['age_x_bmi'] = df['age_proxy'] * df['bmi_proxy']


# ---------------------------------------------------------------------------
print( " ✓ Engineered columns added ✓ " )
display(df.filter([
    'breast','age_menarche_gap','years_since_first_birth','nulliparous','breast_dense_3lvl','breast_dense_binary',
    'menopause_3lvl','menopause_label',
    'age','age_widebin_6lvl','age_widebin_6lvl_label','age_proxy',
    'bmi','is_bmi_missing,''bmi_proxy','years_since_first_birth','density_x_postmeno','postmeno_dense','fam_or_biopsy','density_x_famhx',
]).sample(8))

# ---------------------------------------------------------------------------
```

```python
# Loading Bar
QUIET = True
SEARCH_VERBOSE = 0 if QUIET else 2
EST_VERBOSE    = 0 if QUIET else 1

def fit_with_progress(name, search, X, y, fit_kwargs=None, cv=None, grid=None):
    fit_kwargs = fit_kwargs or {}
    total = None
    if grid is not None and cv is not None:
        total = len(list(ParameterGrid(grid))) * cv.get_n_splits()
    pbar = tqdm(total=total, desc=f"{name} grid", unit="fit", leave=True, dynamic_ncols=True)
    with tqdm_joblib(pbar):
        result = search.fit(X, y, **fit_kwargs)
    pbar.close()
    return result
```

```python
numeric_tf = SkPipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_tf = SkPipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_tf, NUMERIC_COLS),
        ("cat", categorical_tf, CATEGORICAL_COLS),
    ],
    remainder="drop",
    verbose_feature_names_out=False
)
```

```python
sns.set(style="whitegrid"); plt.rcParams["figure.figsize"] = (7.5, 5)
cv = StratifiedKFold(n_splits=CV_FOLDS, shuffle=True, random_state=RANDOM_STATE)

def specificity_score(y_true, y_pred, sample_weight=None):
    cm = confusion_matrix(y_true, y_pred, sample_weight=sample_weight)
    tn, fp, fn, tp = cm.ravel()
    return float(tn) / (tn + fp) if (tn + fp) > 0 else 0.0

def npv_score(y_true, y_pred, sample_weight=None):
    cm = confusion_matrix(y_true, y_pred, sample_weight=sample_weight)
    tn, fp, fn, tp = cm.ravel()
    return float(tn) / (tn + fn) if (tn + fn) > 0 else 0.0

def eval_on_test(name, est, Xt, yt, w=None, thr=0.5):
    proba = est.predict_proba(Xt)[:, 1]
    yhat = (proba >= thr).astype(int)
    cm = confusion_matrix(yt, yhat, sample_weight=w)
    acc = cm.trace() / cm.sum()
    return {
        "model": name,
        "threshold": thr,
        "accuracy": acc,
        "precision": precision_score(yt, yhat, sample_weight=w, zero_division=0),
        "recall": recall_score(yt, yhat, sample_weight=w, zero_division=0),
        "f1": f1_score(yt, yhat, sample_weight=w, zero_division=0),
        "specificity": specificity_score(yt, yhat, sample_weight=w),
        "npv": npv_score(yt, yhat, sample_weight=w),
        "roc_auc": roc_auc_score(yt, proba, sample_weight=w),
        "pr_auc": average_precision_score(yt, proba, sample_weight=w)
    }

def plot_quick_diagnostics(name, est, Xt, yt, w=None, thr=0.5):
    proba = est.predict_proba(Xt)[:, 1]
    yhat = (proba >= thr).astype(int)

    cm = confusion_matrix(yt, yhat, sample_weight=w)
    sns.heatmap(cm, annot=True, fmt=".0f", cmap="Blues")
    plt.title(f"{name} — Confusion Matrix @ {thr:.2f}")
    plt.xlabel("Predicted"); plt.ylabel("Actual"); plt.show()

    fpr, tpr, _ = roc_curve(yt, proba, sample_weight=w)
    auc = roc_auc_score(yt, proba, sample_weight=w)
    plt.plot(fpr, tpr, lw=2, label=f"AUC = {auc:.3f}")
    plt.plot([0,1],[0,1],"--", lw=1); plt.title(f"{name} — ROC Curve")
    plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(); plt.show()

    prec, rec, _ = precision_recall_curve(yt, proba, sample_weight=w)
    ap = average_precision_score(yt, proba, sample_weight=w)
    plt.plot(rec, prec, lw=2, label=f"AP = {ap:.3f}")
    plt.title(f"{name} — Precision–Recall Curve")
    plt.xlabel("Recall"); plt.ylabel("Precision"); plt.legend(); plt.show()
```

```python
def choose_thresholds(est, Xt, yt, w=None,
                      target_recall=0.80, target_precision=0.50, beta=1.0):
    """
    Returns thresholds for best F-beta, recall ≥ target_recall, precision ≥ target_precision.
    Handles PR-curve shapes robustly (P/R are len N+1; thresholds are len N).
    """
    proba = est.predict_proba(Xt)[:, 1]
    P, R, T = precision_recall_curve(yt, proba, sample_weight=w)  # len(P)=len(R)=N+1, len(T)=N
    # Avoid divide-by-zero; compute F-beta for the first N points only (align with T)
    Pn, Rn = P[:-1], R[:-1]  # align with thresholds T
    F = (1 + beta**2) * Pn * Rn / (beta**2 * Pn + Rn + 1e-12)

    # best F-beta (choose argmax over aligned portion)
    i_f = int(np.nanargmax(F)) if F.size else 0
    thr_f = float(T[i_f]) if T.size else 0.5

    # recall ≥ target_recall  (R typically decreases with T; take the largest T that still meets it)
    idx_rec = np.where(Rn >= target_recall)[0]
    thr_rec = float(T[idx_rec[-1]]) if idx_rec.size else (float(T[0]) if T.size else 0.5)

    # precision ≥ target_precision (P often rises with T; take the smallest T that meets it)
    idx_pr = np.where(Pn >= target_precision)[0]
    thr_pr = float(T[idx_pr[0]]) if idx_pr.size else (float(T[-1]) if T.size else 1.0)

    # Clamp to [0,1] just in case
    thr_f   = float(np.clip(thr_f, 0.0, 1.0))
    thr_rec = float(np.clip(thr_rec, 0.0, 1.0))
    thr_pr  = float(np.clip(thr_pr, 0.0, 1.0))

    # Use consistent keys with two decimals for readability
    return {
        "bestF1": thr_f,
        f"recall≥{target_recall:.2f}": thr_rec,
        f"precision≥{target_precision:.2f}": thr_pr
    }

MODEL_VARS = {
    "LogisticRegression": "logreg_best",
    "GradBoost":          "gb_best",
    "HistGradBoost":      "hgb_best",
    "RandomForest":       "rf_best",
    "XGBoost":            "xgb_best",
    "CatBoost":           "cat_best",
    "SVM-RBF":            "svm_best",
    "KNN":                "knn_best",
    "NaiveBayes":         "nb_best",
}
available = {name: globals()[var] for name, var in MODEL_VARS.items() if var in globals()}

target_recall = 0.88
target_precision = 0.70

all_tables, thresholds_store = [], {}
for name, est in available.items():
    th = choose_thresholds(est, X_test, y_test, w=W_test,
                           target_recall=target_recall, target_precision=target_precision, beta=1.0)
    thresholds_store[name] = th

    rows = [
        eval_on_test(f"{name} @0.5", est, X_test, y_test, w=W_test, thr=0.5),
        eval_on_test(f"{name} @bestF1({th['bestF1']:.3f})", est, X_test, y_test, w=W_test, thr=th["bestF1"]),
        eval_on_test(f"{name} @rec≥{target_recall:.2f}({th[f'recall≥{target_recall:.2f}']:.3f})",
                     est, X_test, y_test, w=W_test, thr=th[f"recall≥{target_recall:.2f}"]),
        eval_on_test(f"{name} @prec≥{target_precision:.2f}({th[f'precision≥{target_precision:.2f}']:.3f})",
                     est, X_test, y_test, w=W_test, thr=th[f"precision≥{target_precision:.2f}"])
    ]
    all_tables.append(pd.DataFrame(rows))

ops_df = pd.concat(all_tables, ignore_index=True)
display(ops_df.sort_values(["model","threshold"]).reset_index(drop=True))
```

```python
import numpy as np, joblib
from dataclasses import dataclass

# pick top 3 by PR-AUC (from base_df computed above)
top3_names = base_df["model"].head(3).tolist()
top3_models = [available[n] for n in top3_names]
weights = base_df["pr_auc"].head(3).to_numpy()
weights = weights / weights.sum()

@dataclass
class SoftVotingEnsemble:
    models: list
    weights: np.ndarray
    def predict_proba(self, X):
        ps = [m.predict_proba(X)[:,1] for m in self.models]
        p = np.average(np.vstack(ps), axis=0, weights=self.weights)
        # return 2 columns to mimic predict_proba for binary
        return np.vstack([1-p, p]).T

ensemble = SoftVotingEnsemble(models=top3_models, weights=weights)
ens_res = eval_on_test("Ensemble(top3,weighted)", ensemble, X_test, y_test, w=W_test, thr=0.5)
display(pd.DataFrame([ens_res]))

# tune threshold for ensemble (e.g., best-F1)
from sklearn.metrics import precision_recall_curve
P,R,T = precision_recall_curve(y_test, ensemble.predict_proba(X_test)[:,1], sample_weight=W_test)
F = 2*P*R/(P+R+1e-12); thr_ens = float(T[int(np.nanargmax(F[:-1]))])
ens_res_tuned = eval_on_test(f"Ensemble @bestF1({thr_ens:.3f})", ensemble, X_test, y_test, w=W_test, thr=thr_ens)
display(pd.DataFrame([ens_res_tuned]))
```

# 18) OOF-based Thresholding (Unbiased)

We compute **out-of-fold (OOF) probabilities** on the training set for each tuned model, pick an F1‑optimal threshold on OOF, then evaluate on the hold-out set using that fixed threshold.

```python

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_recall_curve, average_precision_score, roc_auc_score, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_predict
import numpy as np, pandas as pd

def oof_threshold_and_test(estimator_pipeline):
    # OOF probabilities on TRAIN
    oof_proba = cross_val_predict(estimator_pipeline, X_train, y_train, cv=cv, method='predict_proba', n_jobs=-1)[:,1]
    prec, rec, thr = precision_recall_curve(y_train, oof_proba)
    f1 = 2*(prec*rec)/(prec+rec+1e-12)
    i = int(np.nanargmax(f1))
    best_thr = thr[i] if i < len(thr) else 0.5
    # Fit on full TRAIN then evaluate on TEST
    estimator_pipeline.fit(X_train, y_train)
    y_proba_test = estimator_pipeline.predict_proba(X_test)[:,1]
    ap = average_precision_score(y_test, y_proba_test)
    roc = roc_auc_score(y_test, y_proba_test)
    y_pred = (y_proba_test >= best_thr).astype(int)
    cm = confusion_matrix(y_test, y_pred)
    return best_thr, ap, roc, precision_score(y_test, y_pred, zero_division=0), recall_score(y_test, y_pred, zero_division=0), f1_score(y_test, y_pred, zero_division=0), cm

comparison_rows = []
for name, est in fitted_models.items():
    thr, ap, roc, p, r, f1, cm = oof_threshold_and_test(est)
    comparison_rows.append([name, thr, ap, roc, p, r, f1, cm])

comparison_df = pd.DataFrame(comparison_rows, columns=['Model','OOF_best_threshold','Holdout_PR_AUC','Holdout_ROC_AUC','Precision','Recall','F1','ConfusionMatrix']).sort_values('Holdout_PR_AUC', ascending=False)
display(comparison_df)
```

# 23) Model Explainability — SHAP (Global & Local)


We compute SHAP values for the **best model** to understand **which features drive predictions**.
- Uses `shap.TreeExplainer` for tree models (XGBoost/RandomForest/GB).
- If `shap` isn't available in your environment, we fall back to **Permutation Importance** for a global view.


```python

# Pick the best model by Holdout PR-AUC
best_name = combined.sort_values('Holdout_PR_AUC', ascending=False).iloc[0]['Model']
base_name = best_name.split('+')[0]
best_pipe = fitted_models[base_name]
print("Best model selected for explainability:", best_name)

# Fit on training and transform X_test through the preprocessor to get dense matrix & feature names
best_pipe.fit(X_train, y_train)
pre = best_pipe.named_steps['preprocess']
X_test_tr = pre.transform(X_test)

# Feature names
try:
    ohe = pre.named_transformers_['cat'].named_steps.get('ohe', None)
    cat_names = ohe.get_feature_names_out(CATEGORICAL_COLS).tolist() if ohe is not None else []
except Exception:
    cat_names = []
num_names = NUMERIC_COLS
feature_names = num_names + cat_names

# Try SHAP
use_shap = False
try:
    import shap  # may not be installed in this environment
    use_shap = True
except Exception as e:
    print("SHAP not available in this environment; will use permutation importance fallback.", e)

if use_shap:
    try:
        # Extract underlying classifier
        clf = best_pipe.named_steps['clf']
        # Build TreeExplainer (works for XGB, RF, GB)
        explainer = shap.TreeExplainer(clf)
        shap_values = explainer.shap_values(X_test_tr)
        # Global summary (beeswarm)
        shap.summary_plot(shap_values, X_test_tr, feature_names=feature_names, show=True)
    except Exception as e:
        print("SHAP failed; falling back to permutation importance.", e)
        use_shap = False

if not use_shap:
    from sklearn.inspection import permutation_importance
    import numpy as np, matplotlib.pyplot as plt
    r = permutation_importance(best_pipe, X_test, y_test, scoring='average_precision', n_repeats=10, random_state=42, n_jobs=-1)
    imp = r.importances_mean
    order = np.argsort(imp)[::-1][:30]
    labels = np.array(feature_names)[order] if len(feature_names)==len(imp) else np.arange(len(order)).astype(str)
    plt.figure()
    plt.barh(range(len(order))[::-1], imp[order][::-1])
    plt.yticks(range(len(order))[::-1], labels[::-1], fontsize=8)
    plt.xlabel("Permutation Importance (delta AP)")
    plt.title("Top Features — Permutation Importance (Fallback)")
    plt.tight_layout()
    plt.show()
```


# Early Breast Cancer Prediction 

## 1)Setup & Configuration

```python
DATA_PATH = "bcsc_risk_factors.csv"   # Sample / Full dataset
TEST_SIZE = 0.2
CV_FOLDS = 3
RANDOM_STATE = 42
N_JOBS = -1
MAX_PLOT_ROWS = 50000

import sys, platform, importlib, warnings
warnings.filterwarnings("ignore")
print("Python:", sys.version.split()[0])
print("Platform:", platform.platform())

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
                             roc_curve, precision_recall_curve, average_precision_score, ConfusionMatrixDisplay, brier_score_loss)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

try:
    from xgboost import XGBClassifier
    XGB_AVAILABLE = True
except Exception:
    XGB_AVAILABLE = False

pd.set_option("display.max_columns", 180)
pd.set_option("display.width", 220)

print("Config ready. DATA_PATH:", DATA_PATH, 'Yes' if XGB_AVAILABLE else 'No')
```

```python
X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)
```

## 6) Class Imbalance (SMOTE) 

```python
# 2) Build models — all CLASSIFIERS — wrapped as: StandardScaler → SMOTE → clf
clf_lr = LogisticRegression(max_iter=1000, solver="lbfgs")
clf_rf = RandomForestClassifier(n_estimators=400, random_state=RANDOM_STATE, n_jobs=N_JOBS)
clf_gb = GradientBoostingClassifier(random_state=RANDOM_STATE)

pipe_lr = ImbPipeline([("scaler", StandardScaler()), ("smote", SMOTE(random_state=RANDOM_STATE)), ("clf", clf_lr)])
pipe_rf = ImbPipeline([("scaler", StandardScaler()), ("smote", SMOTE(random_state=RANDOM_STATE)), ("clf", clf_rf)])
pipe_gb = ImbPipeline([("scaler", StandardScaler()), ("smote", SMOTE(random_state=RANDOM_STATE)), ("clf", clf_gb)])

models = {"LogReg": pipe_lr, "RandomForest": pipe_rf, "GradBoost": pipe_gb}

if XGB_AVAILABLE:
    clf_xgb = XGBClassifier(
        n_estimators=600, learning_rate=0.05, subsample=0.8, colsample_bytree=0.8,
        max_depth=6, random_state=RANDOM_STATE, n_jobs=N_JOBS, eval_metric="logloss"
    )
    pipe_xgb = ImbPipeline([("scaler", StandardScaler()), ("smote", SMOTE(random_state=RANDOM_STATE)), ("clf", clf_xgb)])
    models["XGBoost"] = pipe_xgb

# Soft-voting ensemble (give Voter plain classifiers; wrap once in a pipeline)
estimators_list = [("lr", clf_lr), ("rf", clf_rf), ("gb", clf_gb)]
if XGB_AVAILABLE:
    estimators_list.append(("xgb", clf_xgb))
voter = VotingClassifier(estimators=estimators_list, voting="soft")
pipe_ens = ImbPipeline([("scaler", StandardScaler()), ("smote", SMOTE(random_state=RANDOM_STATE)), ("clf", voter)])
models["EnsembleSoft"] = pipe_ens

# sanity print
for name, pipe in models.items():
    last = pipe.named_steps["clf"]
    print(f"{name:12s} -> {type(last).__name__}, has predict_proba: {hasattr(last,'predict_proba')}")
```

```python
# 3) Cross‑validation (isolated per model so one failure doesn’t kill all)
scorer = {
    "accuracy": "accuracy",
    "precision": "precision",
    "recall": "recall",
    "f1": "f1",
    "roc_auc": "roc_auc",
    "average_precision": "average_precision",
}
cv = StratifiedKFold(n_splits=CV_FOLDS, shuffle=True, random_state=RANDOM_STATE)

cv_means, cv_folds = {}, {}
for name, pipe in models.items():
    try:
        scores = cross_validate(pipe, X_train, y_train, cv=cv, scoring=scorer,
                                n_jobs=N_JOBS, return_estimator=False, error_score="raise")
        cv_means[name] = {k.replace("test_",""): float(np.mean(v)) for k, v in scores.items() if k.startswith("test_")}
        cv_folds[name] = {k.replace("test_",""): v for k, v in scores.items() if k.startswith("test_")}
        print(f"[OK] {name}")
    except Exception as e:
        print(f"[FAILED] {name}: {e}")

cv_df = pd.DataFrame(cv_means).T.reindex(columns=["roc_auc","average_precision","recall","precision","f1","accuracy"])
display(cv_df)
best_by_auc = cv_df["roc_auc"].idxmax()
print("Best by CV ROC‑AUC:", best_by_auc)
```

```python
# 4) Fit ALL models on train; evaluate on test (so you can compare everything)
test_rows, test_probs = [], {}
for name, pipe in models.items():
    pipe.fit(X_train, y_train)
    proba = pipe.predict_proba(X_test)[:, 1]
    pred = (proba >= 0.5).astype(int)
    test_probs[name] = proba
    test_rows.append({
        "model": name,
        "accuracy": accuracy_score(y_test, pred),
        "precision": precision_score(y_test, pred, zero_division=0),
        "recall": recall_score(y_test, pred, zero_division=0),
        "f1": f1_score(y_test, pred, zero_division=0),
        "roc_auc": roc_auc_score(y_test, proba),
        "average_precision": average_precision_score(y_test, proba),
    })

test_df = pd.DataFrame(test_rows).set_index("model").sort_values("roc_auc", ascending=False)
display(test_df)
```

```python
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
import scipy.stats as st
tuned_results = {}

# -------- Random Forest Tuning --------
rf_base = RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=N_JOBS)
rf_pipe = ImbPipeline([
    ("scaler", StandardScaler()),
    ("smote", SMOTE(random_state=RANDOM_STATE)),
    ("clf", rf_base),
])

rf_dist = {
    "clf__n_estimators": st.randint(400, 1000),
    "clf__max_depth": st.randint(5, 30),
    "clf__min_samples_split": st.randint(2, 20),
    "clf__min_samples_leaf": st.randint(1, 20),
    "clf__max_features": ["sqrt", "log2", None],
}

rf_tuner = RandomizedSearchCV(
    rf_pipe, rf_dist, n_iter=40, scoring="roc_auc",
    cv=cv, random_state=RANDOM_STATE, n_jobs=N_JOBS, verbose=2
)
rf_tuner.fit(X_train, y_train)
print("RF best AUC (CV):", rf_tuner.best_score_)
print("RF best params:", rf_tuner.best_params_)
tuned_results["RF"] = rf_tuner

# -------- XGBoost Tuning (if available) --------
try:
    from xgboost import XGBClassifier
    XGB_AVAILABLE = True
except Exception:
    XGB_AVAILABLE = False

if XGB_AVAILABLE:
    xgb_base = XGBClassifier(
        objective="binary:logistic", eval_metric="logloss", n_jobs=N_JOBS,
        random_state=RANDOM_STATE, tree_method="hist"
    )
    xgb_pipe = ImbPipeline([
        ("scaler", StandardScaler()),
        ("smote", SMOTE(random_state=RANDOM_STATE)),
        ("clf", xgb_base),
    ])

    xgb_dist = {
        "clf__n_estimators": st.randint(500, 1400),
        "clf__max_depth": st.randint(3, 10),
        "clf__learning_rate": st.loguniform(1e-3, 3e-1),
        "clf__subsample": st.uniform(0.6, 0.4),          # 0.6–1.0
        "clf__colsample_bytree": st.uniform(0.6, 0.4),   # 0.6–1.0
        "clf__gamma": st.uniform(0.0, 5.0),
        "clf__reg_alpha": st.uniform(0.0, 1.0),
        "clf__reg_lambda": st.uniform(0.5, 1.5),
    }

    xgb_tuner = RandomizedSearchCV(
        xgb_pipe, xgb_dist, n_iter=60, scoring="roc_auc",
        cv=cv, random_state=RANDOM_STATE, n_jobs=N_JOBS, verbose=2
    )
    xgb_tuner.fit(X_train, y_train)
    print("XGB best AUC (CV):", xgb_tuner.best_score_)
    print("XGB best params:", xgb_tuner.best_params_)
    tuned_results["XGB"] = xgb_tuner
else:
    print("XGBoost not available—skipping tuning.")
```

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, average_precision_score)

tuned_rows = []
tuned_probs = {}

for name, tuner in tuned_results.items():
    best_pipe = tuner.best_estimator_
    proba = best_pipe.predict_proba(X_test)[:, 1]
    pred = (proba >= 0.5).astype(int)
    tuned_probs[name] = proba
    tuned_rows.append({
        "model": f"{name}_TUNED",
        "accuracy": accuracy_score(y_test, pred),
        "precision": precision_score(y_test, pred, zero_division=0),
        "recall": recall_score(y_test, pred, zero_division=0),
        "f1": f1_score(y_test, pred, zero_division=0),
        "roc_auc": roc_auc_score(y_test, proba),
        "average_precision": average_precision_score(y_test, proba),
    })

tuned_df = pd.DataFrame(tuned_rows).set_index("model").sort_values("roc_auc", ascending=False)
print("=== Tuned Models — Test Metrics ===")
display(tuned_df)

# Combine with your previous (untuned) test_df to see gains
combined = pd.concat([test_df, tuned_df], axis=0).sort_values("roc_auc", ascending=False)
print("=== Combined (Untuned + Tuned) — Test Metrics ===")
display(combined)
```

```python
import numpy as np
import matplotlib.pyplot as plt

feature_names = list(X_train.columns)

# ---- Random Forest feature importance ----
if "RF" in tuned_results:
    rf_best = tuned_results["RF"].best_estimator_
    rf_clf = rf_best.named_steps["clf"]
    rf_imp = getattr(rf_clf, "feature_importances_", None)
    if rf_imp is not None:
        topk = min(25, len(rf_imp))
        idx = np.argsort(rf_imp)[-topk:][::-1]
        plt.figure(figsize=(8, 7))
        plt.barh(range(topk), rf_imp[idx][::-1])
        plt.yticks(range(topk), [feature_names[i] for i in idx][::-1])
        plt.title("Random Forest — Top Feature Importances")
        plt.tight_layout(); plt.show()

# ---- XGBoost feature importance ----
if "XGB" in tuned_results:
    xgb_best = tuned_results["XGB"].best_estimator_
    xgb_clf = xgb_best.named_steps["clf"]
    xgb_imp = getattr(xgb_clf, "feature_importances_", None)
    if xgb_imp is not None:
        topk = min(25, len(xgb_imp))
        idx = np.argsort(xgb_imp)[-topk:][::-1]
        plt.figure(figsize=(8, 7))
        plt.barh(range(topk), xgb_imp[idx][::-1])
        plt.yticks(range(topk), [feature_names[i] for i in idx][::-1])
        plt.title("XGBoost — Top Feature Importances")
        plt.tight_layout(); plt.show()
```

```python
import numpy as np
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_curve, precision_recall_curve, brier_score_loss)

# Pick best tuned model by test ROC-AUC
if not tuned_df.empty:
    best_tuned_name = tuned_df.index[0]
    print("Best tuned model by test ROC-AUC:", best_tuned_name)
    best_key = "XGB" if best_tuned_name.startswith("XGB") else "RF"
    best_est = tuned_results[best_key].best_estimator_
    best_proba = tuned_probs[best_key]
else:
    print("No tuned models available; falling back to best untuned.")
    best_tuned_name = test_df.index[0]
    best_est = models[best_tuned_name]
    best_proba = test_probs[best_tuned_name]

# Sweep thresholds
thresholds = np.linspace(0.05, 0.95, 37)
rows = []
for t in thresholds:
    yhat = (best_proba >= t).astype(int)
    rows.append({
        "threshold": t,
        "accuracy": accuracy_score(y_test, yhat),
        "precision": precision_score(y_test, yhat, zero_division=0),
        "recall": recall_score(y_test, yhat, zero_division=0),
        "f1": f1_score(y_test, yhat, zero_division=0),
    })
thresh_tbl = pd.DataFrame(rows)
display(thresh_tbl.head(10))

# Choose thresholds
t_f1 = thresh_tbl.loc[thresh_tbl["f1"].idxmax(), "threshold"]
recall_target = 0.85
candidates = thresh_tbl[thresh_tbl["recall"] >= recall_target]
t_recall = candidates.iloc[np.argmax(candidates["precision"])]["threshold"] if not candidates.empty else t_f1

print(f"Best-F1 threshold: {t_f1:.3f}")
print(f"Recall≥{recall_target:.2f} threshold: {t_recall:.3f}")

# Plot curves
plt.figure()
plt.plot(thresh_tbl["threshold"], thresh_tbl["recall"], label="Recall")
plt.plot(thresh_tbl["threshold"], thresh_tbl["precision"], label="Precision")
plt.plot(thresh_tbl["threshold"], thresh_tbl["f1"], label="F1")
plt.axvline(t_f1, ls="--", label=f"Best F1 @ {t_f1:.2f}")
plt.axvline(t_recall, ls="--", label=f"Recall≥{recall_target:.2f} @ {t_recall:.2f}")
plt.title(f"Metric vs Threshold — {best_tuned_name}")
plt.xlabel("Threshold"); plt.ylabel("Score"); plt.legend(); plt.show()

# Calibration
bins = np.linspace(0.0, 1.0, 11)
digitized = np.digitize(best_proba, bins)-1
xs, ys = [], []
for i in range(len(bins)-1):
    m = digitized == i
    if m.any():
        xs.append(float(np.mean(best_proba[m])))
        ys.append(float(np.mean(y_test.iloc[m])))
plt.figure(); plt.plot(xs, ys, marker="o"); plt.plot([0,1],[0,1],'--')
plt.title(f"Calibration curve — {best_tuned_name}"); plt.xlabel("Avg predicted"); plt.ylabel("Observed pos rate"); plt.show()
print("Brier score:", brier_score_loss(y_test, best_proba))
```
